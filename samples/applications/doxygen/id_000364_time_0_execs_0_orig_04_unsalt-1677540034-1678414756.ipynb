# `unsalt`: remove salt and solvate components

### Introduction

The **`unsalt`** module identifies counterions and solvate components, such that they can be removed. It does not do the removal itself, as each component is put through the whole standardization pipeline before counterions and/or solvate components are identified. Note that any component containing an non-organic element is removed _before_ the standardization pipeline is applied.

### Example

This is a simplified version of how this module is used in the **`standardise`** module. As noted above, components not containing non-organic elements would actually be put through the whole standardization pipeline.

```python
mol = Chem.MolFromSmiles("[Na+].OC(=O)Cc1ccc(CN)cc1.OS(=O)(=O)C(F)(F)F")

Draw.MolsToGridImage(Chem.GetMolFrags(mol, asMols=True))
```

```python
non_salt_frags = []

for frag in Chem.GetMolFrags(mol, asMols=True):
    
    logging.debug("Starting fragment '{}'...".format(Chem.MolToSmiles(frag)))
        
    if unsalt.is_nonorganic(frag): continue # Looks for any 'non-organic' element
        
    if unsalt.is_salt(frag): continue       # Uses salt dictionary
        
    non_salt_frags.append(frag)
    
Draw.MolsToGridImage(non_salt_frags)
```

### The Salts Dictionary

This is used to identify counterions and solvent molecules, and is derived from the ChEMBL salt dictionary. It is stored as a list of (neutral) SMILES.

Recall that bonds to metals are broken, rule-based standardization and charge neutraliszation are all performed before salt matching. Currently, in order to maximise coverage, stereochemistry is removed and matching done on InChIs.

Note that *any* fragment containing a 'non-organic' element (_i.e._ anything not H, C, N, O, S, P, F, Cl, Br or I) is treated as a salt, so some of the entries here are redundant; they are kept for completeness.

```python
salts = pd.DataFrame(list(unsalt.salts.values()))

salts.sort_values("name", inplace=True)

PandasTools.AddMoleculeColumnToFrame(salts, smilesCol='SMILES', molCol='molecule')

pd.set_option('display.max_rows', len(salts))

salts
```
